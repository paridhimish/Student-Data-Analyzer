```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Sample dataset
data = {
    "text": [
        "Government announces new policy",
        "Aliens landed in India yesterday",
        "Stock market rises by 5%",
        "Man claims he can fly",
        "New education reforms introduced"
    ],
    "label": [1, 0, 1, 0, 1]  # 1=Real, 0=Fake
}

df = pd.DataFrame(data)

# Convert text to numbers
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["text"])

y = df["label"]

# Train model
model = LogisticRegression()
model.fit(X, y)

# Prediction
news = input("Enter news: ")
news_vec = vectorizer.transform([news])

pred = model.predict(news_vec)

print("Prediction:", "Real" if pred[0] == 1 else "Fake")
```
