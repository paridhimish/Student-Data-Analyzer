```python


# Student Data Analyzer Project
# Developed using Python, Pandas, Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Taking user input
n = int(input("Enter number of students: "))

names = []
marks = []

for i in range(n):
    name = input("Enter name: ")
    mark = int(input("Enter marks: "))

    names.append(name)
    marks.append(mark)

# Create DataFrame
df = pd.DataFrame({"Name": names, "Marks": marks})

# Basic analysis
print("Average Marks:", df["Marks"].mean())
print("Highest Marks:", df["Marks"].max())
print("Lowest Marks:", df["Marks"].min())

# Topper
topper = df.loc[df["Marks"].idxmax()]
print("Topper Details:\n", topper)

# Pass/Fail
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")

# Grade system
def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    else:
        return "C"

df["Grade"] = df["Marks"].apply(grade)

# Ranking
df["Rank"] = df["Marks"].rank(ascending=False)

# Sorting by rank
df = df.sort_values(by="Rank")

# Improved graph
plt.bar(df["Name"], df["Marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Performance")
plt.show()

# Save file
df.to_csv("student_data.csv", index=False)

df
```
